```python
import  pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv("C:/Users/Admin/Downloads/capetown.csv")
data2=pd.read_csv("C:/Users/Admin/Downloads/johannesburg.csv")
```

```python
data1.head()
```

```python
data2.head()
```

```python
sns.lineplot(data=data1)
plt.title("Cape Town rainfall line graph data")

plt.ylim(0,10)
```

```python
sns.lineplot(data=data2)
plt.title("Johannesburg rainfall line graph data")
plt.ylim(0,10)
```

```python
data1.plot(x="YEAR", y=["JAN", "FEB", "MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"], kind="bar",width=0.8)
plt.title("Cape Town rainfall data")
plt.rcParams["figure.figsize"] = (23,6.5)
```

```python
data2.plot(x="YEAR", y=["JAN", "FEB", "MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"], kind="bar",width=0.8)
plt.title("Johannesburg rainfall data")
plt.rcParams["figure.figsize"] = (23,6.5)
```

```python

plt.pie(data1["ANN"],labels=[2017,2018,2019,2020,2021],explode=[0.2,0,0,0,0],shadow=True)

plt.title("Cape Town Yearly rainfall")
```

```python
plt.pie(data2["ANN"],labels=[2017,2018,2019,2020,2021],explode=[0.2,0,0,0,0],shadow=True)
plt.title("Johannesburg Yearly rainfall")
```
